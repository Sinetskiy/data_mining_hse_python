### Задача 1. (макс. 8 баллов)
Скачайте датасет в .xlsx (Excel) "Археологические находки на реконструируемых улицах" с московского портала открытых данных : https://data.mos.ru/opendata/7702155262-arheologicheskie-nahodki-na-rekonstruiruemyh-ulitsah?pageNumber=2&versionNumber=2&releaseNumber=8

1. (2 балла) Загрузите датасет функцией pandas.read_excel(). Сколько было находок?
2. (1 балла) Добавьте в датафрейм колонки с годом и месяцем обнаружения (метод apply)
4. (1 балл) Выведете распределение количества находок по годам и отдельно по месяцам (метод value_counts())
5. (2 балла) Найдите находку с самым длинным описанием
6. (1 балл) Бонусный балл за интересные факты, которые можно почерпнуть из датасета с использованием pandas.

```python
#1 Загрузите датасет функцией pandas.read_excel(). Сколько было находок?

import pandas as pd 

findings = pd.read_excel(open('data-54581-2018-08-28.xlsx','rb'), sheet_name=0)
len(findings)
```

```python
#2 Добавьте в датафрейм колонки с годом и месяцем обнаружения (метод apply)

findings['Month'] = findings[['DateOfDiscovery']].apply(lambda x: str(x).split('.')[1], axis=1)
findings['Years'] = findings[['DateOfDiscovery']].apply(lambda x: str(x).split('.')[2][:4], axis=1)
findings
```

```python
#3 Выведете распределение количества находок по годам и отдельно по месяцам (метод value_counts())

print(findings['Month'].value_counts())
print(findings['Years'].value_counts())
```

```python
#4 Найдите находку с самым длинным описанием

field_length = findings.DescriptionOfFinding.astype(str).map(len)
print(findings.loc[field_length.idxmax()])
```

```python
#1 Загрузите датасет функцией pandas.read_excel(). Сколько всего площадок?

sports_area = pd.read_excel(open('data-25342-2018-08-17.xlsx','rb'), sheet_name=0)
len(sports_area)
```

```python
#2 Pandas сокращает количество выводимых столбцов, если их слишком много. 
# Выполните pd.set_option('display.max_columns', None) , чтобы выводить всегда все колонки.

pd.set_option('display.max_columns', None) 
sports_area
```

```python
#3 Сколько площадок останется, если удалить строчки, 
# у которых хотя бы 1 значение пропущено (функция pandas.dropna())? 
# Почему столько?

sports_area_not_empty = sports_area.dropna()
sports_area_not_empty
# т.к. хотя бы в одном столбце каждой строчки есть значение NaN 
```

```python
#4  Выведите все бесплатные площадки, приспособленные для инвалидов
sports_area[(sports_area.DisabilityFriendly != 'не приспособлен') & (sports_area.Paid == 'бесплатно')]
```

```python
#5 Какие типы покрытия бывают на площадках (метод unique() у колонки)? 
# Какое покрытие самое популярное, самое непопулярное (метод value_counts())?
for surface in sports_area.SurfaceTypeSummer.unique():
    print(surface)

print('\n##########\n')

surface_type_list = sports_area.SurfaceTypeSummer.value_counts().index.tolist()
print('Самое популярное: {}'.format(surface_type_list[0]))
print('Самое не популярное: {}'.format(surface_type_list[len(surface_type_list)-1]))
```

```python
#6 Сколько всего районов, в которых расположены площадки (метод .nunique() у колонки). 
# Какой район имеет наибольшее количество площадок?
print('Всего районов с площадками: {}'.format(sports_area.District.nunique()))

print('\n##########\n')

spots_are_index = sports_area.District.value_counts().index.tolist()[0];
print('Район имеет наибольшее количество площадок: {} - {}'
      .format(spots_are_index,sports_area.District.value_counts()[spots_are_index]))
```

```python
#7 Сколько площадок в вашем районе? На каком месте находится ваш район по кол-ву площадок? (функция np.where() или метод .index())

for inx, value in enumerate(sports_area.District.value_counts().index.tolist()):
    if(value == 'Даниловский район'):
        print('{} находится на {} месте по количесву площадок'.format(value, inx+1))
    
```
